```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('./mse/training.pickle', 'rb') as f:
    mse_data = pickle.load(f)
```

```python
# results using mse consistency loss (30epochs)
print(mse_data)
```

```python
plt.figure(figsize=(12,6))
plt.plot(mse_data.loss, label='total')
plt.plot(mse_data.class_loss, label='class loss')
plt.plot(mse_data.cons_loss, label='consistency loss')
plt.legend()
plt.show()
```
